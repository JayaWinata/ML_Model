```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
dir(digits)
```

```python
df = pd.DataFrame(digits.data)
df['target'] = digits.target
df
```

GridSearchCV & RandomizedSearchCV is used for find the best parameter values in a ML model

```python
clf = GridSearchCV(svm.SVC(),{
    'C':[1,2,3,4],
    'kernel':['linear', 'poly', 'rbf', 'sigmoid']
},cv=5,return_train_score=False)
clf.fit(digits.data,digits.target)
```

```python
df = pd.DataFrame(clf.cv_results_).sort_values('rank_test_score')
df
```

C = 4, kernel = rbf has the greatest score than the other parameters

Looking for the best model and its best parameter

```python
models = [
    {
        'model': RandomForestClassifier(),
        'params': {
            'n_estimators': [50, 75, 100],
            'criterion': ['gini', 'entropy']
        }
    },
    {
        'model': LogisticRegression(),
        'params': {
            'C': [1, 2, 3]
        }
    },
    {
        'model': GaussianNB(),
        'params': {}
    },
    {
        'model': MultinomialNB(),
        'params': {}
    },
    {
        'model': DecisionTreeClassifier(),
        'params': {
            'criterion': ['gini', 'entropy']
        }
    }
]
```

```python
scores = []
for i in models:
    clf = GridSearchCV(i['model'],i['params'],cv=5,return_train_score=False)
    clf.fit(digits.data,digits.target,)
    scores.append({
        'model':str(i['model']).replace('()',''),
        'best_score':clf.best_score_,
        'best_params':clf.best_params_
    })
```

```python
scores_df = pd.DataFrame(scores)
scores_df.sort_values('best_score',ascending=False)
```
